SegÃºn al contexto y datos proporcionados por el docente y tambien realizada una investigaciÃ³n se dÃ­o a conocer parÃ¡metros Ãºtiles con lo cuales son los siguientes

| **Columna Ãštil**        | **RazÃ³n**                                      |
|-------------------------|-----------------------------------------------|
| `warranty_available`    | La garantÃ­a puede aumentar la confianza del comprador. |
| `free_shipping`         | El envÃ­o gratuito puede influir en la decisiÃ³n de compra. |
| `has_pictures`          | Productos con imÃ¡genes suelen venderse mejor. |
| `num_pictures`          | MÃ¡s imÃ¡genes pueden hacer que el producto sea mÃ¡s atractivo. |
| `log_price`            | Reduce la dispersiÃ³n de precios para mejorar el anÃ¡lisis. |

```python
# Limpieza de datos
# Definir columnas a eliminar
columns_to_drop = ['seller_contact', 'deal_ids', 'non_mercado_pago_payment_methods', 'variations',
                   'coverage_areas', 'descriptions', 'international_delivery_mode', 'video_id',
                   'subtitle', 'catalog_product_id', 'official_store_id', 'differential_pricing']

# AnÃ¡lisis de valores nulos en las columnas antes de eliminarlas
print("\nAnÃ¡lisis de valores nulos antes de eliminar columnas:")
missing_values = df[columns_to_drop].isnull().sum()
total_rows = len(df)

# Calcular porcentaje de valores nulos
missing_percentage = (missing_values / total_rows) * 100

# Crear DataFrame con anÃ¡lisis de nulos
drop_analysis = pd.DataFrame({
    'Column': columns_to_drop,
    'Missing Values': missing_values,
    'Missing %': missing_percentage
})

# Mostrar el anÃ¡lisis antes de eliminar las columnas
print(drop_analysis)

# Eliminar columnas irrelevantes o con demasiados valores nulos
df.drop(columns=columns_to_drop, inplace=True)

# Convertir fechas a formato datetime
date_columns = ['date_created', 'last_updated', 'start_time', 'stop_time']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Extraer caracterÃ­sticas Ãºtiles
df['warranty_available'] = df['warranty'].notnull()
df['free_shipping'] = df['shipping'].apply(lambda x: x.get('free_shipping', False) if isinstance(x, dict) else False)
df['has_pictures'] = df['pictures'].apply(lambda x: len(x) > 0 if isinstance(x, list) else False)
df['num_pictures'] = df['pictures'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df['log_price'] = np.log1p(df['price'])

# SelecciÃ³n de caracterÃ­sticas relevantes para el modelo
selected_features = ['log_price', 'warranty_available', 'free_shipping', 'has_pictures', 'num_pictures', 'condition']
df_model = df[selected_features].dropna()
```

```python
# Guardar el dataset limpio
df_model.to_csv('/content/drive/MyDrive/MaestriaETL/data_clean.csv', index=False)

print("Dataset limpio guardado con Ã©xito.")
```
